#Reading csv files


```python
df = (spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv("/Volumes/workspace/bronze/source_systems/source_crm/sales_details.csv"))
```

write it to bronze layer

```python
df.write.mode("overwrite").saveAsTable("bronze.crm_sales_details")
```
